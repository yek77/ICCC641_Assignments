```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft

# -----------------------
# Generate simple signal
# -----------------------

fs = 8000  # lower sample rate for clearer plotting
duration = 3.0
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

signal = np.zeros_like(t)
signal[t < 1.5] = np.sin(2 * np.pi * 300 * t[t < 1.5])
signal[t >= 1.5] = np.sin(2 * np.pi * 1500 * t[t >= 1.5])

# -----------------------
# Plot 1: Zoomed waveform
# -----------------------

plt.figure()
plt.plot(t[:1000], signal[:1000])  # zoom first 1000 samples
plt.xlabel("Time (seconds)")
plt.title("Zoomed Waveform (First 1000 samples)")
plt.show()
```
